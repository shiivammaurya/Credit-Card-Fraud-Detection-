```python
#under -samplling
#build a sample dataset containing similiar dataset as normal transaction and the fraudlent transaction
```

here we are going to take 191 fraud and 191 legit transaction and make a samll sample dataset


```python
legit_sample= legit.sample(n=191)
#now concatanating to make a small dataset
new_dataset=pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
x=new_dataset.drop(columns='Class',axis=1)
y=new_dataset['Class']
```

```python
print(x)
```

splitting the data into training and testing data

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

```python
print(x_train)
```

```python
#model training Logistic regression from sklearn model
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
 #accruacy on training data
 x_train_prediction=model.predict(x_train)
 training_data_accuracy=accuracy_score(x_train_prediction,y_train)
```

```python
print('Accuracy on training data :',training_data_accuracy)
```
